```python
# 판다스 라이브러리 생성(셀마다 지정하고 싶을 때)
import pandas as pd
# 데이터 생성
list1 = list([
    ['홍길동', '남자', 30, 183],
    ['지은탁', '여자', 24, 162],
    ['한서준', '남자', 23, 172],
    ['이수호', '남자', 21, 182],
    ['임주경', '여자', 17, 155],
    ['이두나', '여자', 26, 157],
    ['강남순', '여자', 24, 157],
    ['도봉순', '여자', 24, 172],
])
col_names = ['이름','성별','나이','키']

# 데이터 프레임으로 생성 후 저장
df=pd.DataFrame(list1, columns=col_names)

# 출력
df
```

```python
# 키가 180보다 큰 사람만 조회
df[df['키']>170]
# 조건식을 정수형으로 비교할 때 데이터가 정수형일때만 비교하여 조회 가능
```

### 조건식에 isin()함수, and기호(&), or기호(|), str.contains()함수를 사용하여 더 복잡한 조건으로 조회할 수도 있다

```python
# 나이가 21세 또는 23세인 사람을 조회
df[df['나이'].isin([21,23])]
```

```python
# 성별이 '여자'이면서 키가 160보다 큰 사람을 조회
# and => &
df[(df['성별']=='여자')&(df['키']>160)]
```

```python
# 나이가 28세 이상이거나 성별이 '남자'인 사람을 조회
# or => |
df[(df['나이']>=28)|(df['성별']=='남자')]
```

```python
# 이름에 '순'이 들어간 사람을 조회
# .str.contains(str)
df[df['이름'].str.contains('순')]
```

```python
# 나이와 키에 대한 통계 출력
# 숫자형 데이터만 통계 출력
df.describe()
```

```python
# 인덱스 변경
# inplace=True => 원본 데이터프레임에 적용
df.set_index('이름', inplace=True)
# 한번만 실행 가능 (이미 이름이 Key로 적용이 되었기 때문에 오류발생)
df
```

# 성별을 기준으로 조회하여 키의 평균을 조회

```python
df.describe()

# 평균 => mean(), 표준편차 => std()
```

```python
# 성별의 평균 키
# 성별을 기준으로 그룹화하여 그룹별 키의 평균을 계산
# as_index=False => 데이터프레임으로 출력 됨
mean_by_gender = df.groupby(by=['성별'], as_index=False)['키'].mean()

# 열 이름 바꾸기
mean_by_gender.rename(columns={'키':'평균키'}, inplace=True)

# 출력
mean_by_gender
```

```python
# 성별의 키의 표쥰편차
std_by_gender=df.groupby(by=['성별'], as_index=False)['키'].std()

# 열 이름 바꾸기
std_by_gender.rename(columns={'키':'키의 표준 편차'}, inplace=True)

# 출력
std_by_gender
```

```python
# 두 데이터 프레임 합치기
new_df=pd.merge(mean_by_gender,std_by_gender)

# 출력
new_df
```

```python
# 열 이름과 시간표 데이터를 리스트로 저장
col_names=['과목번호', '과목명', '강의실', '시간수']
list1=list([
    ['C1', '파이썬', '303호', 3],
    ['C2', '국어', '201호', 3],
    ['C3', '영어', '203호', 3],
    ['C4', '한국사', '201호', 2],
    ['C5', '과학', '401호', 3],
    ['C6', '사회', '201호', 2]])
```

```python
# 시간표 데이터를 데이터프레임 객체 df로 변환하여 CSV 파일로 저장
import pandas as pd

# 데이터 프레임으로 변환
timetable_df=pd.DataFrame(list1, columns=col_names)

# 데이터 프레임을 CSV 파일로 저장
timetable_df.to_csv('./timetable.csv', header=True, index=False, encoding='utf-8')

# CSV 파일을 읽어와서 ","(쉼표)로 구분하여 저장
timetable_df = pd.read_csv('./timetable.csv', sep=',')

# 출력
timetable_df
```

```python
# 열 이름 '담당 교사' 추가
# 값 -> '이순신', '김유신', '강감찬', '신사임당', '정약용', '허준'
timetable_df['담당 교사']=['이순신', '김유신', '강감찬', '신사임당', '정약용', '허준']

# 출력
timetable_df
```

```python
# 강의실별 최대 시간 구하기
max_hour = timetable_df.groupby(by=['강의실'], as_index=False)['시간수'].max()

# 출력
max_hour
```
